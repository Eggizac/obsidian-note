У этой системы бесконечно много решний
Тут вычисление частных решений через общее. Общее решение выражается через параметры y = t, z = s

```python
import numpy as np

A = np.array([[1,1,1],
              [2,2,2]], dtype=float)
b = np.array([1,2], dtype=float)

def sol(t, s):
    x = 1 - t - s
    return np.array([x, t, s], dtype=float)

for t, s in [(0,0), (1,0), (0.2, -3)]:
    x = sol(t, s)
    print("t,s=", t, s, "Ax=", A@x, "ok=", np.allclose(A@x, b))
```

```python
# Эта функция считает скалярное произведение
def dot(a ,b):
    if len(a) != len(b):
        raise ValueError("Vectors have the some lenght")
    return sum(x*y for x, y in zip(a, b))

a = np.array([2, -1], dtype=float)
b = np.array([5, 3], dtype = float)

print(dot(a, b))
```

```python
# Нахождение косинуса
def cosine_similarity(x, y):
    x = np.asanyarray(x).reshape(-1)
    y = np.asanyarray(y).reshape(-1)

    norm_x = np.linalg.norm(x)
    if norm_x == 0:
        raise ValueError("x it's a null vector")
    norm_y = np.linalg.norm(y)
    if norm_y == 0:
        raise ValueError("y it's a null vector")
    
    dot_xy = dot(x, y)

    cosine = (dot_xy / (norm_x * norm_y))

    return cosine

x = np.array([1, 2, 3], dtype=float)      # shape (3,)
y = np.array([[1], [2], [3]], dtype=float) # shape (3,1)

print(cosine_similarity(x, y))
```
