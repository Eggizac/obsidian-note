Реализация линейной регрессии на питоне

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'x':[0.9, 4.11, 4.82, 8.53, 8.9], 'y':[1.91, 1.84, 4.47, 6.38, 3.92]})
```

```python
df
```

```python
df.plot(x = 'x', y = 'y',kind='scatter')
```

```python
m,b = 1,1
```

```python
df['y_pred'] = df['x'] * m + b
```

```python
df
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df['x'], df['y'])
plt.plot(df['x'], df['y_pred'])
```

```python
df['err_abs'] = np.abs(df['y'] - df['y_pred'])

df['err_abs'].mean()
```

```python
df['m_change'] = df['x'] * np.sign(df['y'] - df['y_pred'])
```

```python
df['b_change'] = np.sign(df['y'] - df['y_pred'])

print(df['b_change'].mean())
```
